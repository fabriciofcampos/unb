# Funções utilizadas.

Coloque abaixo as funções desenvolvidas para este projeto.

```python
#Código das funções utilizadas.
def rect():
    x = np.linspace(-2, 2, 32769)
    for i in range(len(x)):
        if abs(x[i]) > 0.5:
            x[i] = 0
        elif abs(x[i]) == 0.5:
            x[i] = 0.5
        else:
            x[i] = 1
    return np.array(x)


def sinc(W):
    x = np.linspace(-8* np.pi, 8* np.pi, 32769)
    for i in range(len(x)):
        if not x[i]:
            x[i] = 1
            continue
        x[i] = np.sin(x[i]/W) / (x[i]/W)
    return np.array(x)


def amostra(sinal, T):
    x_amostrado = list()
    n = 0
    while n < 32:
        c = 0
        for t in np.linspace(0, 4, 32769):
            c = sinal[int(n*T*8192)]
        n += 1
        x_amostrado.append(c)
    return np.array(x_amostrado)


def DFT(N, sinal):
    WN = np.exp(-1j * 2*np.pi/(N))
    X = list()
    for i in range(N):
        c = 0
        for j in range(N):
            c += sinal[j] * np.power(WN, (i-16) * j)
        X.append(c)
    return np.array(X)


def H1(N):
    h = list()
    fc = 1/16
    for n in range(N):
        if n < fc*N or n > N - (fc*N):
            h.append(1)
            continue
        elif n == fc*N or n == N - (fc*N):
            h.append(0.5)
            continue
        h.append(0)
    return h


def frequencia(espectro):
    espectro_mag = abs(espectro)
    espectro_max = sorted(espectro_mag, reverse=True)
    frequencias = list()
    for i in range(len(espectro)):
        if espectro_mag[i] in espectro_max[:10]:
            frequencias.append(i)
    return frequencias


def teclas(sinal, N):
    return len(sinal) // N


def identifica_numero(sinal, N):
    tabelado = [697, 770, 852, 940, 1209, 1336, 1477]
    somas = [2276, 1906, 2033, 2174, 1979, 2106, 2247, 2061, 2188, 2329]
    n_teclas = teclas(sinal, N)
    sinal1s = np.array_split(sinal, n_teclas)
    numero = list()
    for sinal in sinal1s:
        freqs = list()
        espectro = np.fft.fft(sinal, n=N)
        frequencias = frequencia(espectro)
        for f in frequencias:
            if f in tabelado:
                freqs.append(f)
        numero.append(somas.index(sum(freqs)))
    return numero
```

```python
#Resposta da questão 2a
N = 256
x = list()
c = N
while c:
    c -= 1
    if N - c < 65:
        x.append(1)
        continue
    x.append(0)
X = np.fft.fft(x)

Y = H1(N) * X
y = np.fft.ifft(Y)

plt.plot(x, c='blue')
plt.plot(y, c='orange')

plt.title('Sinais x[n] e y[n]')
plt.xlabel('n')
plt.ylabel('Sinal')
plt.show()
```
